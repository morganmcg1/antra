```python
from pathlib import Path
```

```python
path = Path('data/irish/parallel_corpora/tatoeba')
```

```python
import pandas as pd
sent_details = pd.read_csv(path/'sentences_detailed.csv', 
                           sep='\t', header=None, error_bad_lines=False)
sent_details = sent_details.rename(columns={0:'Sentence id', 1:'Lang',  2:'Text', 
                                            3:'Username', 4:'Date added', 
                                            5:'Date last modified'})
sent_details.head()
```

```python
links = pd.read_csv(path/'links.csv', sep='\t', header=None, error_bad_lines=False)
links = links.rename(columns={0:'Sentence id',1:'Translation id'})
links.head()
```

```python
sent_details.loc[sent_details.Lang == 'gle'].merge()
```

```python
dd = pd.merge(sent_details.loc[sent_details.Lang == 'gle'], links, left_on='Sentence id', right_on='Sentence id')
```

```python
dd.head()
```

```python
ee = pd.merge(dd, sent_details, left_on='Translation id', right_on='Sentence id')
df_fin = ee[['Sentence id_x','Text_x','Sentence id_y','Lang_y','Text_y']]
df_fin = df_fin.loc[df_fin.Lang_y == 'eng']
df_fin.columns = ['ga_id', 'ga', 'en_id', 'en_l', 'en']
df_fin = df_fin[['ga_id', 'ga', 'en_id', 'en']]
print(len(df_fin))
df_fin.head()
```

```python
df_fin.to_csv(path/'tatoeba_en-ga.csv', index=False)
```
